```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_rows', None)
df_list = []
for year in range(2017, 2022, 2):
    excelstring=f"Defense{year}.csv"
    year_df = pd.read_csv(excelstring)
    df_list.append(year_df)

df = pd.concat(df_list, ignore_index=True, axis=0)
display(df.head()) 
```

First, creating the same master data frame from 2017, 2019, and 2021 from before.

```python
top_shifters_RH = df.loc[df['RHH - %'] > 25.0]
top_shifters_LH = df.loc[df['LHH - %'] > 25.0]
overall_top_shifters = df.loc[df['%'] > 25.0]
overall_top_shifters.head()
```

Creating the same dataFrames for teams that shift most frequently.

```python
ranked_top_shifters = overall_top_shifters.sort_values(by=['%'], ascending=False)
ranked_top_shifters.head()
```

```python
worst_shifters = overall_top_shifters.sort_values(by=['%'], ascending=True)
worst_shifters.head()
```

From these two dataFrames, we see that the team over the three years covered in this table was the 2021 Los Angeles Dodgers, while we see that the team that shifted the least often over the same time span was the 2021 San Francisco Giants. Interestingly, these two teams had the second-best and best records, respectively, across MLB for the 2021 season, despite their dramatically different approaches to tech defense.
